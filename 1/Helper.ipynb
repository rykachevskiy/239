Стоимость проезда в электричке от расстояния 
Стоимость квартиры в зависимости от площади

```python
X = np.array([30, 32, 35, 37, 40, 41, 45, 47, 48,49, 53, 57, 60, 61,62,68,70, 75, 78, 80])
y = np.log(np.log(np.log(X)))* 15000 + (np.random.random(X.shape) - 0.5)  * 500
#y = 1/(1+np.exp(-X/ 15)) * 
y_ = np.log(np.log(np.log(X)))* 15000
```

```python
plt.figure(figsize = (10, 7))
font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 16}

matplotlib.rc('font', **font)
plt.plot(X, y, markersize = 15, marker = 'x', linewidth=0)
plt.plot(X, y_, linewidth=2, color = 'g')
plt.xlabel("Size, m^2")
plt.ylabel("Price, 1000rub")

```
